# Customer Churn Prediction - Feature Engineering

## Objective

The objective of this notebook is to prepare the cleaned customer churn dataset for machine learning. This includes selecting relevant features, encoding categorical variables, handling numerical features, and creating the final dataset required for model training.

## Business Problem

Customer churn is one of the biggest challenges for telecom companies. Losing customers leads to revenue loss and increased acquisition costs. The goal of feature engineering is to transform the cleaned customer data into a machine learning-ready dataset that helps predict whether a customer is likely to churn.
